### Data Dictionary

FEATURES: 
- Title: The title of the book 
- Author: The author(s) of the book. 
- Edition: The edition of the book eg (Paperback,– Import, 26 Apr 2018) 
- Reviews: The customer reviews about the book 
- Ratings: The customer ratings of the book 
- Synopsis: The synopsis of the book 
- Genre: The genre the book belongs to
- BookCategory: The department the book is usually available at. 
- Price: The price of the book (Target variable)

###### Data Preprocessing

```python
print(f'Null Values : \n{data.isnull().sum()}\n')
print(f'Dupliate Values : \n{data.duplicated().sum()}') # ambiguous
```

```python
data['Reviews'] = data['Reviews'].str.split(' ', expand = True)[0].astype(float)
```

```python
data['Ratings'] = data['Ratings'].str.split(' ', expand = True)[0].apply(lambda x : int(x) if len(x)<3 
                                                       else int(x.replace(',','')))
```

```python
data['EditionYear'] = data['Edition'].str.split(' ').apply(lambda x : x[-1]).apply(lambda x : int(x) 
                                                                                   if x.isdigit() 
                                                                                   else 0)
# some zeros present.
```

```python
def preprocess_text(text):
    # lowercasing
    lowercased_text = text.lower()
    
    # cleaning
    remove_anomalies = re.sub(r'[\n|.]',' ', lowercased_text)
    remove_punctuation = re.sub(r'[^\w\s]','', remove_anomalies)
    remove_white_spaces = remove_punctuation.strip()
    
    # tokenization - breaking the word down into series of array
    tokenized_text = nltk.tokenize.word_tokenize(remove_white_spaces)
    
    # StopWords/filtering - a, and, the removals
    stop_words = set(nltk.corpus.stopwords.words('english'))
    stopwords_removed = [word for word in tokenized_text if word not in stop_words]
    
    # stremming - transfering words into their base form
    ps = nltk.stem.PorterStemmer()
    stemmed_text = [ps.stem(word) for word in stopwords_removed]
    
    return stemmed_text
```

```python
data['Genre'] = data['Genre'].apply(preprocess_text)
data['BookCategory'] = data['BookCategory'].apply(preprocess_text)
data['Synopsis'] = data['Synopsis'].apply(preprocess_text)
```

```python
data['Author_Name'] = data['Author'].apply(lambda x : re.split('/s', x.lower().replace(' ','')))
```

```python
data['tag1'] = data['Synopsis'] + data['Genre'] + data['BookCategory'] + data['Author_Name']
data['tag2'] = data['Synopsis'] + data['Genre'] + data['BookCategory'] 
```

```python
edition_type = {'Softcover Editions': ['Paperback', 'Mass Market Paperback', 'Tankobon Softcover', 'Perfect Paperback', 
        '(Kannada),Paperback', '(German),Paperback', '(French),Paperback', 
        '(Spanish),Paperback', '(Chinese),Paperback', 'Flexibound'],
    'Hardcover Editions': ['Hardcover', 'Library Binding', 'Leather Bound'],
    'Non-standard Formats': ['Sheet music', 'Plastic Comb', 'Loose Leaf', 'Board book', 'Cards', 
        'Spiral-bound', 'Product Bundle']}
flattened_edition_type = {edition: category for category, editions in 
                          edition_type.items() for edition in editions}    
data['Edition_Type'] = data['Edition'].str.split(',–', expand = True)[0]
data['Edition_Type'] = data['Edition_Type'].map(flattened_edition_type)
```

```python
sns.boxplot(x = data['Edition_Type'], y = data['Price']);
# yes, indeed it is true.. Hardcovers are more costlier than Softcovers.
```

```python
data.drop(columns = ['Author','Edition','Synopsis','Genre'], inplace = True)
```

```python
data['BookCategory'] = data['BookCategory'].apply(lambda x : ' '.join(x))
```

```python
data['tag1'] = data['tag1'].apply(lambda x : ' '.join(x))
data['tag2'] = data['tag2'].apply(lambda x : ' '.join(x))
```

```python
# train_test split
train = data.iloc[0:train_data.shape[0]]
test = data.iloc[train_data.shape[0]:]
```

```python
map_dict = train.groupby(by = 'BookCategory')['Price'].median().to_dict()
train['BookCategory'] = train['BookCategory'].map(map_dict)
test['BookCategory'] = test['BookCategory'].map(map_dict)
```

```python
# possibilty of applying of content based clustering...
```

```python
train_copy = pd.get_dummies(train.drop(columns = ['Title','Author_Name','tag1','tag2']), drop_first = True)
test_copy = pd.get_dummies(test.drop(columns = ['Title','Author_Name','tag1','tag2']), drop_first = True)
```

```python
x_train = train_copy.drop('Price', axis = 1)
y_train = train_copy['Price']

x_test = test_copy.drop('Price', axis = 1)
```

```python
test_copy.isnull().sum()
```

```python
def custom_metric(ytrue,ypred): # provided by machine hack
    log_diff = np.log10(ypred + 1) - np.log10(ytrue + 1)
    rmsle = np.sqrt(np.square(log_diff).mean())
    metric_value = 1 - rmsle
    return metric_value
```

```python
dtree = DecisionTreeRegressor()
y_pred = dtree.fit(x_train,y_train).predict(x_train)
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred,y_train))}')
print(f'R2_score: {r2_score(y_pred,y_train)}')
print(f'RMSLE: {custom_metric(y_pred,y_train)}')
```

```python
rm = RandomForestRegressor()
y_pred = rm.fit(x_train,y_train).predict(x_train)
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred,y_train))}')
print(f'R2_score: {r2_score(y_pred,y_train)}')
print(f'RMSLE: {custom_metric(y_pred,y_train)}')
```

```python
xgb = XGBRegressor()
y_pred = xgb.fit(x_train,y_train).predict(x_train)
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred,y_train))}')
print(f'R2_score: {r2_score(y_pred,y_train)}')
print(f'RMSLE: {custom_metric(y_pred,y_train)}')
```

```python
gbm = GradientBoostingRegressor() # poor
y_pred = gbm.fit(x_train,y_train).predict(x_train)
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred,y_train))}')
print(f'R2_score: {r2_score(y_pred,y_train)}')
print(f'RMSLE: {custom_metric(y_pred,y_train)}')
```

```python
ada = AdaBoostRegressor() # poor
y_pred = ada.fit(x_train,y_train).predict(x_train)
print(f'RMSE: {np.sqrt(mean_squared_error(y_pred,y_train))}')
print(f'R2_score: {r2_score(y_pred,y_train)}')
print(f'RMSLE: {custom_metric(y_pred,y_train)}')
```
